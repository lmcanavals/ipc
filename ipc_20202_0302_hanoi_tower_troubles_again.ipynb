```
%%writefile hanoi.cpp
#include <bits/stdc++.h>

using namespace std;

int main() {
    int T, N, i, k = 0;
    cin >> T;

    while (T--) {
        cin >> N;
        vector<int> x(N, 0);
        for (i = 1; k < N;++i) {
            for (k = 0; k < N; ++k) {
                int y = sqrt(x[k]+i);
                if (x[k] == 0 || y*y == x[k]+i) {
                    x[k] = i;
                    break;
                }
            }
        }
        cout << i-2 << endl;
    }
}
```

```
%%writefile hanoi2.cpp
#include <bits/stdc++.h>

using namespace std;

int main() {
    int T, N, i, k = 0;
    cin >> T;

    while (T--) {
        cin >> N;
        vector<int> T(N+1, 0);
        T[1] = 1;
        for (int i = 2; i < N+1; ++i) {
            T[i] = T[i-1]+i + (i % 2 == 0? 0 : 1);
        }
        cout << T[N] << endl;
    }
}
```
